```python

def pearson_similarity(user1,user2,ratingmatrix_for_training,cols1,rows1,averages1=0,averages2=0,n1=0,n2=0):
  #  print("hello")
  #  print(ratingmatrix_for_training)
    i1=0
    flag1=0
    flag2=0
    
    while(i1<cols1):
        averages2+=ratingmatrix_for_training[user2][i1]
        if(ratingmatrix_for_training[user2][i1]>0):
            n2+=1
            flag1=1
        i1+=1
    i1=0
    while(i1<cols1):
        averages1+=ratingmatrix_for_training[user1][i1]
        if(ratingmatrix_for_training[user1][i1]>0):
            n1+=1
            flag2=1
        i1+=1
    if flag1!=0:
   #     print("there1")
        averages2/=n2;

    if flag2!=0:
   #     print("there2")
        averages1/=n1;
    j1=0
    sum=0
    square1=0
    square2=0
    while(j1<=cols1-1):
        if(ratingmatrix_for_training[user1][j1]>0):
            square1+=(averages1-ratingmatrix_for_training[user1][j1])*(averages1-ratingmatrix_for_training[user1][j1])
        if(ratingmatrix_for_training[user2][j1]>0):
            square2+=(averages2-ratingmatrix_for_training[user2][j1])*(averages2-ratingmatrix_for_training[user2][j1])
        if((ratingmatrix_for_training[user2][j1]>0) and (ratingmatrix_for_training[user1][j1]>0) ):
            sum+=(averages1-ratingmatrix_for_training[user1][j1])*(averages2-ratingmatrix_for_training[user2][j1])
        j1+=1
  #  print("helloooo")
    square1=math.sqrt(square1)
    square2=math.sqrt(square2)
    if square1==0 and square2==0:
        #print("here1")
        return 1
    elif square1==0:
      #  print("here2")
        return 0
    elif square2==0:
        #print("here3")
        return 0
   
    return sum/(square1*square2)
        
            
        
            
                                                                                                                              
```

```python
def cosine_similarity(user1,user2,ratingmatrix_for_training,cols1,rows1,averages1=0,averages2=0,n1=0,n2=0):
    i12=0
    sum1=0
    sum2=0
    num=0
    while(i12<cols1):
        if ratingmatrix_for_training[user1][i12]>0 and ratingmatrix_for_training[user2][i12]>0:
            num+=ratingmatrix_for_training[user1][i12]*ratingmatrix_for_training[user2][i12]
        sum1+=ratingmatrix_for_training[user1][i12]*ratingmatrix_for_training[user1][i12]
        sum2+=ratingmatrix_for_training[user2][i12]*ratingmatrix_for_training[user2][i12]
        i12+=1
    if sum1==0 and sum2==0:
        return 1
    elif sum1==0:
        return 0
    elif sum2==0:
        return 0
    sum1=math.sqrt(sum1)
    sum2=math.sqrt(sum2)
    ans=num/(sum1*sum2)
    return ans
     
    
```
